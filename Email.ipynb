```python
#find frequency of the 1,2,3,4,5,6 letters in datastet

variables = EmailData.columns
one_letter = []
two_letters = []
three_letters = []
four_letters =[]
five_letters = []
six_letters = []

for col in variables:
    if len(col) == 1:
        one_letter.append(col)
    if len(col) == 2:
        two_letters.append(col)
    if len(col) == 3:
        three_letters.append(col)
    if len(col) == 4:
        four_letters.append(col)
    if len(col) == 5:
        five_letters.append(col)
    if len(col) == 6:
        six_letters.append(col)





data_one_letter = EmailData[one_letter]
data_two_letter = EmailData[two_letters]
data_three_letter = EmailData[three_letters]
data_four_letter = EmailData[four_letters]
data_five_letter = EmailData[five_letters]
data_six_letter = EmailData[six_letters]





list[data_one_letter.shape, data_two_letter.shape, data_three_letter.shape, data_four_letter.shape, data_five_letter.shape, data_six_letter.shape]
```

```python
#find frequency of the 10,11,12,13,14,15 letters in datastet


variables = EmailData.columns
fifteen_letter = []
fourteen_letters = []
thirteen_letters = []
twelve_letters = []
eleven_letters = []
ten_letters = []

for col in variables:
    if len(col) == 10:
        ten_letters.append(col)
    if len(col) == 11:
        eleven_letters.append(col)
    if len(col) == 12:
        twelve_letters.append(col)
    if len(col) == 13:
        thirteen_letters.append(col)
    if len(col) == 14:
        fourteen_letters.append(col)
    if len(col) == 15:
        fifteen_letter.append(col)


data_ten_letter = EmailData[ten_letters]
data_eleven_letter = EmailData[eleven_letters]
data_twelve_letter = EmailData[twelve_letters]
data_thirteen_letter = EmailData[thirteen_letters]
data_fourteen_letter = EmailData[fourteen_letters]
data_fifteen_letter = EmailData[fifteen_letter]


list[data_ten_letter.shape, data_eleven_letter.shape, data_twelve_letter.shape,
     data_thirteen_letter.shape, data_fourteen_letter.shape, data_fifteen_letter.shape]
```
